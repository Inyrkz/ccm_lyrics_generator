## Transform

```python
def parse_text(lyrics_text):
    '''Function to clean text before appending to the file.
    The function will remove the Verse, title, chorus
    
    Parameters
    ----------
    lyrics_text: str, the lyrics to clean
    
    
    Return
    -------
    cleaned lyrics file
    '''
    try:        
        # remove title from lyrics, it is the first item
        lyrics_text = lyrics_text.split('\n')  # convert to list
        if len(lyrics_text) != 0:  # make sure list isn't empty
            lyrics_text.pop(0)
        # remove the extra space
        if len(lyrics_text) != 0:
            lyrics_text.pop(0)
        lyrics_text = '\n'.join(lyrics_text)
        # remove tags
        lyrics_text = lyrics_text.replace('[Verse]', '')
        lyrics_text = lyrics_text.replace('[Verse 1]\n', '')
        lyrics_text = lyrics_text.replace('[Verse 2]\n', '')
        lyrics_text = lyrics_text.replace('[Verse 3]\n', '')
        lyrics_text = lyrics_text.replace('[Verse 4]\n', '')
        lyrics_text = lyrics_text.replace('[Verse 5]\n', '')
        lyrics_text = lyrics_text.replace('[Verse One]\n', '')
        lyrics_text = lyrics_text.replace('[Verse Two]\n', '')
        lyrics_text = lyrics_text.replace('[Verse Three]\n', '')
        lyrics_text = lyrics_text.replace('[Verse Four]\n', '')
        lyrics_text = lyrics_text.replace('[Verse Five]\n', '')
        lyrics_text = lyrics_text.replace('[Instrumental]\n', '')
        lyrics_text = lyrics_text.replace('[Post-Chorus]\n', '')
        lyrics_text = lyrics_text.replace('[Post-Chorus 1]\n', '')
        lyrics_text = lyrics_text.replace('[Post-Chorus 2]\n', '')
        lyrics_text = lyrics_text.replace('[Post-Chorus 3]\n', '')
        lyrics_text = lyrics_text.replace('[Post-Chorus 4]\n', '')
        lyrics_text = lyrics_text.replace('[Post-Chorus 5]\n', '')
        lyrics_text = lyrics_text.replace('[Chorus]\n', '')
        lyrics_text = lyrics_text.replace('[Chorus 1]\n', '')
        lyrics_text = lyrics_text.replace('[Chorus 2]\n', '')
        lyrics_text = lyrics_text.replace('[Chorus 3]\n', '')
        lyrics_text = lyrics_text.replace('[Chorus 4]\n', '')
        lyrics_text = lyrics_text.replace('[Chorus 5]\n', '')
        lyrics_text = lyrics_text.replace('[Outro]\n', '')
        lyrics_text = lyrics_text.replace('[Bridge 1]\n', '')
        lyrics_text = lyrics_text.replace('[Bridge 2]\n', '')
        lyrics_text = lyrics_text.replace('[Bridge 3]\n', '')
        lyrics_text = lyrics_text.replace('[Bridge 4]\n', '')
        lyrics_text = lyrics_text.replace('[Bridge]\n', '')
        lyrics_text = lyrics_text.replace('[Instrumentals]\n', '')
        lyrics_text = lyrics_text.replace('[Pre-Chorus]\n', '')
        lyrics_text = lyrics_text.replace('[Pre-Chorus 1]\n', '')
        lyrics_text = lyrics_text.replace('[Pre-Chorus 2]\n', '')
        lyrics_text = lyrics_text.replace('[Pre-Chorus 3]\n', '')
        lyrics_text = lyrics_text.replace('[Pre-Chorus 4]\n', '')
        lyrics_text = lyrics_text.replace('[Interlude]\n', '')
        lyrics_text = lyrics_text.replace('[Spontaneous]\n', '')
        lyrics_text = lyrics_text.replace('Chorus:', '')
        lyrics_text = lyrics_text.replace('Verse:', '')
        lyrics_text = lyrics_text.replace('Bridge:', '')
        lyrics_text = lyrics_text.replace('Jonathan David', '')
        lyrics_text = lyrics_text.replace('Melissa Helser', '')
        lyrics_text = lyrics_text.replace('[Spoken]', '')
        lyrics_text = lyrics_text.replace('Embed', '\n')
        
                               
                            
    # if there is an index error, catch it
    except IndexError as e:
        print(e)
        
    return lyrics_text    
```

## Load

```python
# loop through all the json files
for filename in glob.glob(os.path.join('lyrics_raw', '*.json')):
    # read each JSON lyric file
    with open(filename, 'r') as file:
        # open file to load text to
        with open('ccm-full-lyrics.txt', 'a', encoding="utf-8") as f:
            # load lyrics json file
            lyrics = json.load(file)
            # loop through each song in the file
            # to extract lyrics and load to text file
            for song_lyric in lyrics.get('songs'):
                new_text = song_lyric.get('lyrics')
                # clean text with the parse_text() function
                new_text = parse_text(new_text)
                # append to file
                f.write(new_text)
```
